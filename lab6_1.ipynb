```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD

import pandas as pd2
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('C:\projects\git\PetResearch\wine.csv')
X = df.iloc[:,1:12]
y = df.iloc[:,0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from keras.utils import np_utils

y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
y_train = y_train[:,1:] # метод to_categorical выше конвертирует классы которые нужно предсказать в бинарную матрицу. Почему-то добавляет лишний нулевой столбец в начало матрицы, его-то и удаляю.
y_test = y_test[:,1:]
```

```python
type(y_test)
```

```python
X_train = X_train.values
X_test = X_test.values
```

```python
model = Sequential()

model.add(Dense(512,activation='relu',input_dim=11)) # relu - функция активации нейрона,
# позволяет создавать большие нейронные сети которые требуют меньше вычислительных мощностей, relu(x): 0 если x<0;  x если x>=0
model.add(Dropout(0.2)) # дропаут - просто с некоторой вероятностью (0.01) выключаем какой-либо нейрон (вид регуляризации)
model.add(Dense(512,activation='relu')) # 254 -число нейронов в скрытом слое (2 шт.)
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu')) # 254 -число нейронов в скрытом слое (2 шт.)
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu')) # 254 -число нейронов в скрытом слое (2 шт.)
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu')) # 254 -число нейронов в скрытом слое (2 шт.)
model.add(Dropout(0.2))
#model.add(Dense(512,activation='relu')) # 254 -число нейронов в скрытом слое (2 шт.)
#model.add(Dropout(0.2))
model.add(Dense(3,activation='softmax')) # softmax - ф-я активации, используется на выходном слое в мультиклассовой
# классификации. Позволяет принимать решения о классе более "уверенно". Если просто, то чем выше (более вероятным) 
# будет одно из предсказанных значений (классов), тем ниже будут вероятности у остальных классов.
```

```python
# SGD - стохастический градиентный спуск
# lr (learning rate) - скорость обучения, чем меньше значение, тем медленнее обучение(шаг градиента), тем проще попасть в 
                            # локальный минимум (в точку экстремума функции). Вообще его обычно делают динамическим (параметр nesterov).

# decay - указывает, как lr меняется с эпохой.
# momentum - Параметр, ускоряющий SGD в соответствующем направлении и гасящий колебания.
# nesterov - влияет на скорость обучения, если кратко, то это адаптивный метод градиентого спуска, который меняет параметры
# спуска в зависимости от происходящего с функцией (учитывает "ландшафт" функции).
sgd = SGD(lr=0.0001,momentum=0.9,nesterov=True) # decay=1e-6
```

```python
# loss - Вычисляет потери кроссэнтропии между метками и предсказаниями для многоклассовой классификации.(ниже в выводе в консоль) | categorical_crossentropy - функция потерь
# accuracy - точность (коэффициент детерминации R^2)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=1800,verbose=1)
```

```python
rez = model.evaluate(X_test,y_test)
rez_t = model.evaluate(X_train,y_train)

# функция потерь | точность (R^2)  результаты > 90% 
print('loss: ', rez_t[0])
print('acc:  ', rez_t[1])
print('============================')
print('loss: ', rez[0])
print('acc:  ', rez[1])

# TODO + кроссвалидация
```
