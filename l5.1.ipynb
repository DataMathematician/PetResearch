```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# берем данные
sale_price = pd.read_csv('D:\\maks\\lab5\\sample_submission.csv')
test = pd.read_csv('D:\\maks\\lab5\\test.csv')
train = pd.read_csv('D:\\maks\\lab5\\train.csv')
```

```python
test['SalePrice'] = sale_price['SalePrice'].astype(dtype=int)
```

```python
test.info()
train.info()
```

```python
data = pd.concat([train,test],ignore_index=True).drop(['Id'],axis = 1)
```

```python
data = pd.get_dummies(data)
```

```python
# Замена nan на численные данные. Число, на которое меняем nan - зависит от интерпритации данных, попрой неожиданных
# LotFrontage - исходя из признака, можно предположить, что если значение nan, то прилегающей территории нету, значит 0
data['LotFrontage'].fillna(0,inplace=True)

# MasVnrArea площадь облицовки отделочными материалами, раз nan то облицовки нету - пусть будет 0
data['MasVnrArea'].fillna(0,inplace=True)

# BsmtFinSF1 - нет подвала (все что связано ниже с подвалом - встречается только у наблюдения id:2121 или 2120 если считать по индексам, вообще говоря, такое наблюдение можно и удалить)
data['BsmtFinSF1'].fillna(0,inplace=True)

# BsmtFinSF2
data['BsmtFinSF2'].fillna(0,inplace=True)

# BsmtUnfSF
data['BsmtUnfSF'].fillna(0,inplace=True)

# TotalBsmtSF 
data['TotalBsmtSF'].fillna(0,inplace=True)

# BsmtFullBath - тоже 0
data['BsmtFullBath'].fillna(0,inplace=True)

#BsmtHalfBath - тоже 0, принадлежит одному и тому же наблюдению, что и nan в BsmtFullBath
data['BsmtHalfBath'].fillna(0,inplace=True)

# GarageYrBlt - если nan то гараж не строился (тут похожая ситуация с id:2121)
data['GarageYrBlt'].fillna(0,inplace=True) # это может интерпретироваться как 0 год постройки, хотя он не построен, об этом говорят другие признаки свазанные с гаражом

# GarageCars - если nan то забыли указать, наверно, в любом случае - гаража нету
data['GarageCars'].fillna(0,inplace=True)

# GarageArea - тоже самое
data['GarageArea'].fillna(0,inplace=True)
```

```python
# Адекватная, недообученная модель
from sklearn.model_selection import train_test_split 
y = data['SalePrice']
X = data.iloc[:,:].drop('SalePrice',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.499, shuffle = True)
```

```python
# Неадекватная модель
y_train = data.loc[:1459,'SalePrice']
X_train = data.loc[:1459,:].drop('SalePrice',axis=1)

y_test = data.loc[1460:,'SalePrice']
X_test = data.loc[1460:,:].drop('SalePrice',axis=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train)
```

```python
model.score(X_train,y_train) #R^2
```

```python
model.score(X_test,y_test)
```

```python
prices_new = model.predict(X_test)
```

```python
# MSE для предсказанных значений
from sklearn.metrics import mean_squared_error
print('MSE:', np.sqrt(mean_squared_error(y_test,prices_new)))
```

```python
# MSE для тестовых значений
center = np.mean(y_test)
print('MSE:', np.sqrt(sum([(x-center)**2 for x in y_test])/ len(y_test)))
```
